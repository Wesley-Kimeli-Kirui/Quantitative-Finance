#Lists

```python
# Creating a list
fruits = ['apple', 'banana', 'cherry']
numbers = [1, 2, 3, 4, 5]
mixed = ['hello', 42, True]

# Accessing list items
print(fruits[0])  # Output: apple
print(numbers[3])  # Output: 4

# Modifying list items
fruits[1] = 'orange'
print(fruits)  # Output: ['apple', 'orange', 'cherry']

# List concatenation
combined_list = fruits + numbers
print(combined_list)  # Output: ['apple', 'orange', 'cherry', 1, 2, 3, 4, 5]

# List slicing
first_three = numbers[:3]
print(first_three)  # Output: [1, 2, 3]

last_two = numbers[-2:]
print(last_two)  # Output: [4, 5]

# List methods
fruits.append('grape')  # Add an item to the end of the list
print(fruits)  # Output: ['apple', 'orange', 'cherry', 'grape']

fruits.insert(1, 'mango')  # Insert an item at a specific index
print(fruits)  # Output: ['apple', 'mango', 'orange', 'cherry', 'grape']

fruits.remove('orange')  # Remove the first occurrence of an item
print(fruits)  # Output: ['apple', 'mango', 'cherry', 'grape']

fruits.pop(2)  # Remove and return the item at a specific index
print(fruits)  # Output: ['apple', 'mango', 'grape']

print(len(fruits))  # Output: 3 (get the length of the list)

fruits.sort()  # Sort the list in ascending order
print(fruits)  # Output: ['apple', 'grape', 'mango']

reversed_fruits = sorted(fruits, reverse=True)  # Create a new sorted list in descending order
print(reversed_fruits)  # Output: ['mango', 'grape', 'apple']
```

#Integers/Floats

```python
# Integers
a = 10
b = -5

# Integer operations
print(a + b)  # Output: 5 (addition)
print(a - b)  # Output: 15 (subtraction)
print(a * b)  # Output: -50 (multiplication)
print(a / b)  # Output: -2.0 (division, result is a float)
print(a // b)  # Output: -2 (floor division, result is an int)
print(a % b)  # Output: 0 (modulus, remainder of division)
print(a ** b)  # Output: -100000 (exponentiation)

# Integer methods
print(abs(-10))  # Output: 10 (absolute value)
print(pow(2, 3))  # Output: 8 (power function)
print(divmod(10, 3))  # Output: (3, 1) (quotient and remainder of division)

# Floats
c = 3.14
d = 2.5

# Float operations
print(c + d)  # Output: 5.64
print(c - d)  # Output: 0.6399999999999997
print(c * d)  # Output: 7.85
print(c / d)  # Output: 1.256

# Type conversion
e = int(c)  # Convert float to int
print(e)  # Output: 3

f = float(a)  # Convert int to float
print(f)  # Output: 10.0

# Built-in functions
print(round(c, 2))  # Output: 3.14 (round to 2 decimal places)
print(max(a, b, c))  # Output: 10.0 (maximum value)
print(min(a, b, c))  # Output: -5 (minimum value)

# Type checking
print(isinstance(a, int))  # Output: True
print(isinstance(c, float))  # Output: True
```

#Tuples

```python
# Creating a tuple
fruits = ('apple', 'banana', 'cherry')
numbers = (1, 2, 3, 4, 5)
mixed = ('hello', 42, True)

# Accessing tuple items
print(fruits[0])  # Output: apple
print(numbers[3])  # Output: 4

# Tuples are immutable (cannot be modified)
# fruits[1] = 'orange'  # This will raise a TypeError

# Tuple concatenation
combined_tuple = fruits + numbers
print(combined_tuple)  # Output: ('apple', 'banana', 'cherry', 1, 2, 3, 4, 5)

# Tuple slicing
first_three = numbers[:3]
print(first_three)  # Output: (1, 2, 3)

last_two = numbers[-2:]
print(last_two)  # Output: (4, 5)

# Tuple methods
print(len(fruits))  # Output: 3 (get the length of the tuple)

print(fruits.count('apple'))  # Output: 1 (count occurrences of an item)
print(fruits.index('banana'))  # Output: 1 (get the index of an item)

# Tuples are immutable, but you can create a new tuple with modified values
new_fruits = fruits + ('grape',)
print(new_fruits)  # Output: ('apple', 'banana', 'cherry', 'grape')

# Tuples can be unpacked
fruit1, fruit2, fruit3 = fruits
print(fruit1)  # Output: apple
print(fruit2)  # Output: banana
print(fruit3)  # Output: cherry
```

#Dictionaries

```python
# Creating a dictionary
person = {'name': 'Alice', 'age': 30, 'city': 'New York'}
scores = {'math': 85, 'science': 92, 'english': 88}

# Accessing dictionary values
print(person['name'])  # Output: Alice
print(scores['math'])  # Output: 85

# Modifying dictionary values
person['age'] = 31  # Update an existing key-value pair
person['email'] = 'alice@example.com'  # Add a new key-value pair
print(person)  # Output: {'name': 'Alice', 'age': 31, 'city': 'New York', 'email': 'alice@example.com'}

# Removing dictionary items
del person['city']  # Remove a key-value pair
print(person)  # Output: {'name': 'Alice', 'age': 31, 'email': 'alice@example.com'}

# Checking if a key exists
print('name' in person)  # Output: True
print('occupation' in person)  # Output: False

# Getting dictionary keys and values
print(person.keys())  # Output: dict_keys(['name', 'age', 'email'])
print(person.values())  # Output: dict_values(['Alice', 31, 'alice@example.com'])

# Iterating over a dictionary
for key, value in person.items():
    print(f"{key}: {value}")
# Output:
# name: Alice
# age: 31
# email: alice@example.com

# Dictionary methods
print(len(person))  # Output: 3 (get the number of key-value pairs)
person.clear()  # Remove all key-value pairs
print(person)  # Output: {}
```

#Functions

```python
# Defining a simple function
def greet(name):
    """
    This function greets a person by name.
    """
    print(f"Hello, {name}!")

# Calling the function
greet("Alice")  # Output: Hello, Alice!
greet("Bob")  # Output: Hello, Bob!

# Function with parameters and return value
def add_numbers(a, b):
    """
    This function adds two numbers and returns the result.
    """
    result = a + b
    return result

sum1 = add_numbers(2, 3)
sum2 = add_numbers(4, 5)
print(sum1)  # Output: 5
print(sum2)  # Output: 9

# Function with default parameter value
def greet_with_default(name, message="Good morning"):
    """
    This function greets a person with a default or custom message.
    """
    print(f"{message}, {name}!")

greet_with_default("Alice")  # Output: Good morning, Alice!
greet_with_default("Bob", "Hello")  # Output: Hello, Bob!

# Function with arbitrary number of arguments
def sum_numbers(*args):
    """
    This function calculates the sum of an arbitrary number of arguments.
    """
    total = 0
    for arg in args:
        total += arg
    return total

result1 = sum_numbers(1, 2, 3)  # Output: 6
result2 = sum_numbers(4, 5, 6, 7)  # Output: 22
print(result1)
print(result2)
```

#Object-Oriented Programming

```python
# Defining a class
class Vehicle:
    def __init__(self, make, model, year):
        """
        Initialize the Vehicle object.
        """
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        """
        Display information about the vehicle.
        """
        print(f"Make: {self.make}, Model: {self.model}, Year: {self.year}")

# Creating objects (instances of the class)
car1 = Vehicle("Toyota", "Camry", 2020)
car2 = Vehicle("Honda", "Civic", 2018)

# Accessing object attributes and methods
car1.display_info()  # Output: Make: Toyota, Model: Camry, Year: 2020
print(car2.model)  # Output: Civic

# Inheritance
class ElectricVehicle(Vehicle):
    def __init__(self, make, model, year, battery_capacity):
        """
        Initialize the ElectricVehicle object.
        """
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity

    def display_info(self):
        """
        Display information about the electric vehicle.
        """
        super().display_info()
        print(f"Battery Capacity: {self.battery_capacity} kWh")

# Creating an object of the derived class
ev = ElectricVehicle("Tesla", "Model S", 2022, 100)
ev.display_info()
# Output:
# Make: Tesla, Model: Model S, Year: 2022
# Battery Capacity: 100 kWh

# Polymorphism
vehicles = [car1, car2, ev]
for vehicle in vehicles:
    vehicle.display_info()
    print("-" * 20)
```

#ARIMA Modeling on Financial Data:

below you will learn how to utilize the ARIMA model to predict on time series data by utilizing [statsmodels](https://www.statsmodels.org/stable/index.html) & [sklearn](https://scikit-learn.org/stable/)

```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
import warnings
warnings.filterwarnings('ignore')
```

```python
# loading AAPL data into a dataframe

stock = "AAPL"
start = "2015-01-01"
end = "2024-01-01"

df = yf.download(stock, start=start, end=end)
df.isnull().values.any()
df = df.dropna()

df.head()
```

#Train & Test Data

```python
train_data = df.iloc[0:int(len(data)*0.8), :]
test_data = df.iloc[int(len(data)*0.8):, :]

fig = go.Figure()

fig.add_trace(go.Scatter(x=train_data.index, y=train_data['Open'], mode='lines', name='Training Data', line=dict(color='blue')))
fig.add_trace(go.Scatter(x=test_data.index, y=test_data['Open'], mode='lines', name='Testing Data', line=dict(color='green')))

fig.update_layout(
    title='AAPL Open Prices: Training and Testing Data',
    xaxis_title='Dates',
    yaxis_title='Open Prices',
    template='plotly_white'
)

fig.show()
```

```python
window = 7
train_series = train_data['Open']

#Determing rolling statistics
rolmean = train_series.rolling(window).mean()
rolstd = train_series.rolling(window).std()

#Plot rolling statistics:
fig = plt.figure(figsize=(10, 6))
orig = plt.plot(train_series, color='blue',label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation');
```

Setting up an [Augmented Dickey-Fuller test](https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test) to dtermine if the data is stationary

```python
from statsmodels.tsa.stattools import adfuller

window = 7
train_series = train_data['Close']

dftest = adfuller(train_series, autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
dfoutput
```

you can see that the p-value is close to 1, so we fail to reject null hypothesis that the data is not stationary

The stationarity of time series data is crucial because most of the models for forecasting assume that the underlying data are stationary

```python
# calculates the difference between consecutive elements and dropping any na values

train_diff = train_series.diff(periods=1)
train_diff = train_diff.dropna(inplace = False)
```

```python
# determine rolling statistics
window = 7

rolling_mean = train_diff.rolling(window).mean()
rolling_std = train_diff.rolling(window).std()

# plot rolling statistics:
fig = plt.figure(figsize=(10, 6))
orig = plt.plot(train_diff, color='blue',label='Original')
mean = plt.plot(rolling_mean, color='red', label='New Rolling Mean')
std = plt.plot(rolling_std, color='black', label = 'New Rolling Std')
plt.legend(loc='best')
plt.title('New Rolling Mean & Standard Deviation');
```

retrying the [Augmented Dickey-Fuller test](https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test) to determine if the data is stationary

```python
dftest = adfuller(train_diff, autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
dfoutput
```

Now you can see that the p-value of 1.18 ^ -14 is less than our statistically significant threshold of 0.05 (In other words, the p-value is less than one-tenth of one ten-millionth (0.0000000001)).  We can now reject the null hypothesis that the data is not stationary and proceed.  

```python
# Symmetric Mean Absolute Percentage Error (SMAPE)

def smape_kun(y_true, y_pred, epsilon=1e-8):
    denominator = (np.abs(y_pred) + np.abs(y_true)) + epsilon
    absolute_percentage_error = np.abs(y_pred - y_true) * 200 / denominator
    mean_smape = np.mean(absolute_percentage_error)
    return mean_smape
```

```python
test_series = test_data['Close']
test_diff = test_series.diff(periods=1)
test_diff = test_diff.dropna(inplace = False)
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Initialize history with training data
history = [x for x in train_diff]
predictions = list()

# Iterate through the test data points
for t in range(len(test_diff)):
    p, d, q = 5,1,0
    model = ARIMA(history, order=(p, d, q))
    model_fit = model.fit()

    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)

    obs = test_diff[t]
    history.append(obs)

    if t % 100 == 0:
        print('Test Series Point: {}\tPredicted={}, Expected={}'.format(t, yhat, obs))

# Calculate Mean Squared Error (MSE) to evaluate model performance
mse = mean_squared_error(test_diff, predictions)
print('Mean Squared Error (MSE):', mse)
```
